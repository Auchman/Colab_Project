```python
# prompt: create a dataframe that has midterm grades for 100 students

import pandas as pd
import numpy as np

# Create a list of student IDs
student_ids = [f"Student_{i}" for i in range(1, 101)]

# Generate random midterm grades between 60 and 100
midterm_grades = np.random.randint(60, 101, size=100)

# Create a DataFrame
df = pd.DataFrame({"StudentID": student_ids, "MidtermGrade": midterm_grades})

# Print the DataFrame
df
```

```python
# @title MidtermGrade

from matplotlib import pyplot as plt
df['MidtermGrade'].plot(kind='hist', bins=20, title='MidtermGrade')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
